````python
#preparação para o exercício

url_sp = 'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=SP&tipo_interessado=Estado'
url_rj = 'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=RJ&tipo_interessado=Estado'
````

````python
sp = requests.get(url_sp)
rj = requests.get(url_rj)
````

````python
print(data_json_sp['items'])
````

comentário: podemos ver que as duas bases tem as mesmas informações quando comparamos os tipos de dados, que seriam as informações que elas contém, porém a base de São Paulo tem algumas linhas de informações a mais.

````python
sao_paulo_sp['uf'].value_counts()
````

sendo assim, vamos unir as bases para trabalhar com uma só, contendo as informações de São Paulo e Rio de Janeiro

````python
df_sp_e_rj = pd.merge(sao_paulo_sp, rio_de_janeiro_rj, how='outer')
df_sp_e_rj
````

````python
#vamos contar a quantidade UF apenas para sabermos se não foi pedida alguma informação
df_sp_e_rj['uf'].value_counts()
````

````python
#1) Para fazer a função pedida no exercício, primeiro terei que baixar a base com todos os estados.
#tentei por diversas vezes e o site trava, vou então baixar separado as bases de Bahia e Minas Gerais e unir com as de SP e RJ
url_ba = 'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=BA&tipo_interessado=Estado'
url_mg = 'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=MG&tipo_interessado=Estado'
````

````python
ba = requests.get(url_ba)
mg = requests.get(url_mg)
````

````python
data_json_ba = ba.json()
data_json_mg = mg.json()
````

````python
print(data_json_ba['items'])
````

````python
print(data_json_mg['items'])
````

````python
bahia_ba = pd.DataFrame(data_json_ba['items'])
````

````python
minas_gerais_mg = pd.DataFrame(data_json_mg['items'])
minas_gerais_mg
````

````python
df_ba_e_mg = pd.merge(bahia_ba, minas_gerais_mg, how='outer')

df_consolidada = pd.merge(df_ba_e_mg, df_sp_e_rj, how='outer')

df_consolidada['uf'].value_counts()
````

````python
# 2) Seu código aqui
#Arquivado por decurso de prazo de Minas Gerais

decurso_prazo_mg = df_consolidada[(df_consolidada.uf=='MG') & (df_consolidada.status=='Arquivado por decurso de prazo')].count()[0]
decurso_prazo_mg
````

````python
# 3) Seu código aqui
#Qual é o município da Bahia com mais solicitações deferidas?
#nosso Dataframe está escrito por estado e não município. Teremos que ajustar um novo df.
url_ba_municipio = 'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=BA&tipo_interessado=Munic%C3%ADpio'
ba_municipio = requests.get(url_ba_municipio)
data_json_ba_municipio = ba_municipio.json()
bahia_ba_municipio = pd.DataFrame(data_json_ba_municipio['items'])
bahia_ba_municipio
````

````python
#agora com a base da bahia por município, vamos verficar as solicitações deferidas.
#separei então na 'deferidos_ba', todos os status deferidos
deferidos_ba = bahia_ba_municipio[(bahia_ba_municipio.status=='Deferido')]
#agora vamos contar os municípios
deferidos_ba['interessado'].value_counts()
````

````python
# 4) Seu código aqui
#Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'
interessados_ba = df_consolidada[(df_consolidada.interessado=='Bahia')]
interessados_ba.to_csv('interessados_ba.csv')
````
